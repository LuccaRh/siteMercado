```python
import pandas as pd
import numpy as np
import pyodbc
```

```python
df = pd.read_csv(r"SpoonacularApiDatasetSimples.csv", delimiter = ",")
df.head()
```

```python
#Gerar números aleatórios para representar quantidade de produtos
df['Quantidade'] = np.random.randint(10, 31, size=len(df))
#multiplicar por 100 unidades pequenas
df.loc[df['Unidade'].isin(['gram', 'drop','piece','apricot']), 'Quantidade'] *= 10
df.head()
```

```python
#tentei fazer utilizando bibliotecas, mas não é muito preciso
def traducao(unidade):
    if unidade == 'gram':
        return 'gramas'
    elif unidade in ['squash', 'apricot', 'small']:
        return 'unidade'
    elif unidade == 'strip':
        return 'tira'
    elif unidade == 'drop':
        return 'gota'
    elif unidade == 'handful':
        return 'punhado'
    elif unidade == 'piece':
        return 'pedaço'
    elif unidade == 'package':
        return 'pacote'
    elif unidade == 'can':
        return 'lata'
    
df['Unidade'] = df['Unidade'].apply(lambda x: traducao(x))
df.head()
```

```python
conn = pyodbc.connect('Driver={SQL Server};'
                      'Server=LUCCA-NOTEBOOK;'
                      'Database=siteMercadoDB;'
                      'Trusted_Connection=yes;')

cursor = conn.cursor()
```

```python
TableColumns = "Nome, Valor, Unidade, Quantidade, Imagem"
for index, linha in df.iterrows():
    values_tuple = (
        linha['Nome'], linha['Valor R$'], linha['Unidade'], linha['Quantidade'], linha['image']
    )
    cursor.execute(f"INSERT INTO ProdutosTB ({TableColumns}) VALUES (?, ?, ?, ?, ?)", values_tuple)

cursor.commit()
```
